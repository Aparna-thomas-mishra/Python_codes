```python
# Loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns #Data visualization library using matplotlib
```

```python
# Loading the ipl matches dataset
csv_path= '/content/sample_data/matches_final.csv'
ipl=pd.read_csv(csv_path)
```

```python
# view first 10 records of the dataset
ipl.head(10)
```

```python
# View the number of rows and columns in the dataset

ipl.shape
```

```python
# Getting the frequency of player of the match awards
ipl['player_of_match'].value_counts()
```

```python
# Top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
list(ipl['player_of_match'].value_counts()[0:10].keys()) #would give only the names of the players
```

```python
# Make a bar-plot for top 10 players with most man of the match awards
list(ipl['player_of_match'].value_counts()[0:10].keys()) #would give only the names of the players

plt.figure(figsize=(12,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:10].keys()), list(ipl['player_of_match'].value_counts()[0:10]),color='g')
plt.show()
```

```python
# Getting the frequency of results column

ipl['result'].value_counts()
```

```python
# find which team has won the most number of toss

ipl['toss_winner'].value_counts()[0:1]
```

```python
# find teams which have won the match by batting first

batting_first= ipl[ipl['win_by_runs']!=0]
batting_first.head(5)
```

```python
# make histogram. find frequency of runs of matches where the first batting team has won

plt.figure(figsize=(10,5))
plt.hist(batting_first['win_by_runs'],bins=50)
# bins are the number of buckets created for the values provided i.e. The height of each bin shows how many values from that data fall into that range.
plt.title("Distribution of runs")
plt.xlabel("Runs scored")
plt.ylabel("frequency")
plt.show()
```

```python
# make histogram. find frequency of runs of matches where the first batting team has won

plt.figure(figsize=(10,5))
plt.hist(batting_first['win_by_runs'],bins=20)
plt.title("Distribution of runs")
plt.xlabel("Runs scored")
plt.show()
```

```python
# Find out the number of wins wrt to batting first

batting_first['winner'].value_counts()
```

```python
# Make bar-plot for top 5 teams with most wins after batting first

plt.figure(figsize=(13,5))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()), list(batting_first['winner'].value_counts()[0:5]), color=["blue","yellow","red","orange","green"])
plt.show()
```

```python
# Make a pie chart of above

plt.figure(figsize=(5,5))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.2f%%')
# autopct='%0.1f%%'- used for showing percentage distribution in pie chart
plt.show()
```
