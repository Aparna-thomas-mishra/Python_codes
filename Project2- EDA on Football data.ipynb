```python
# Import required libraries

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Read csv file

fifa=pd.read_csv('/content/sample_data/fifa data.csv')
```

```python
# View first 5 data in the dataset

fifa.head()
```

```python
# find the (no. of records, no. of columns) in the dataset

fifa.shape
```

```python
# Find column headers in the dataset

for column in fifa.columns:
  print(column)
```

```python
# Find number of players from each country

fifa['nationality'].value_counts()
```

```python
# Find top ten countries where most players belong to

fifa['nationality'].value_counts()[0:10]
```

```python
#find names of top 5 countries where most of the players come from

fifa['nationality'].value_counts()[0:5].keys()
```

```python
# create bar graph of top 5 countries with most number of players

plt.figure(figsize=(10,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()), list(fifa['nationality'].value_counts()[0:5]), color='pink')
plt.show()
```

```python
# find top 5 earning players

player_salary=fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
# sort the above data in ascending order

player_sal=player_salary.sort_values(by=['wage_eur'],ascending=False)
player_sal.head()
```

```python
# Create bar graph for the above table created

plt.figure(figsize=(10,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color='purple')
plt.xlabel('player names')
plt.ylabel('salary in euro')
plt.show()
```

```python
# Find all players from Germany

ger_players= fifa[fifa['nationality']=='Germany']
ger_players.head(10)
```

```python
# find the tallest German player

ger_players.sort_values(by=['height_cm'],ascending=False).head(1)
```

```python
#find the German player with highest weight

ger_players.sort_values(by=['weight_kg'],ascending=False).head()
```

```python
# find the highest paid German player

ger_players.sort_values(by=['wage_eur'],ascending=False).head()
```

```python
# Find the highest paid German player (only name and wage columnn required)

ger_players[['short_name','wage_eur']].sort_values(by=['wage_eur'], ascending=False).head(5)
```

```python
# find player with best shooting skills

player_shooting=fifa[['short_name','shooting']]
player_shooting.sort_values(by=['shooting'], ascending=False).head()
```

```python
# Find player with best defense skills

player_defending=fifa[['short_name','defending','nationality','club']]
player_defending.sort_values(by=['defending'],ascending=False).head()
```

```python
# Sort players from Real madrid by salary

real_madrid=fifa[fifa['club']=='Real Madrid']
real_madrid=real_madrid[['short_name','wage_eur']]
real_madrid.sort_values(by=['wage_eur'],ascending=False).head()
```
